1.Discrete Custom ColorMap

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from matplotlib.colors import ListedColormap
```

```python
# Create a simple dataset
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)
```

```python
# Define custom discrete colors
colors = ["darkblue", "blue", "green", "yellow", "orange", "red"]
cmap = ListedColormap(colors)
```

```python
plt.figure(figsize=(6, 5))
```

```python
plt.contourf(X, Y, Z, levels=20, cmap=cmap)
```

```python
plt.colorbar()
```

```python
plt.title("Custom Listed Colormap")
```

```python
plt.show()
```

2.Continuous Gradient Custom Colormap

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Create a simple dataset
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)
```

```python
# Define a gradient colormap (blue → white → red)
colors = ["blue", "white", "red"]
cmap = LinearSegmentedColormap.from_list("custom_gradient", colors, N=256)
```

```python
plt.figure(figsize=(6, 5))
```

```python
plt.contourf(X, Y, Z, levels=100, cmap=cmap)
```

```python
plt.colorbar()
```

```python
plt.title("Custom Gradient Colormap")
```

```python
plt.show()
```
